```r
x=1:10
cat(x)
cat('Mean of x is',mean(x)) 
cat('Median of x is',median(x))# cat and print both can be used for printing
cat('Varience of x is', var(x))
print(paste('standard deviation of x', sd(x))) #for print we have to use paste() command in case of string in the print.
```

```r
x=1:10
print(x)
print(paste('Mean of x is',mean(x)))
print(paste('Median of x is',median(x)))
print(paste('Varience of x is', var(x)))
print(paste('standard deviation of x', sd(x))) #for print we have to use paste() command in case of string in the print.
```

```r
x=1:10
cat(x,'\n')
cat('Mean of x is',mean(x),'\n') 
cat('Median of x is',median(x),'\n')
cat('Varience of x is', var(x),'\n')
cat('standard deviation of x', sd(x),'\n')
```
